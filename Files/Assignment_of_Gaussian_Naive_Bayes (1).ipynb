## Guassian Distribution

$$
f(x,\mu,\sigma) = \frac{1}{\sigma \sqrt{2\pi} } e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
$$

```python
def normal_dist(x , mean , sd):
      prob_density = (1/sd*np.sqrt(2*np.pi)) * np.exp(-0.5*((x-mean)/sd)**2)
      return prob_density
```

### Our Goal:

$$
P(Iris|sepal \enspace length, sepal \enspace width, petal \enspace length, petal \enspace width)
$$

You will need to create the **y_pred**

```python
pd.unique(y['class'])
```

```python
# class-wise mean and sd for each feature

classes = list(pd.unique(y_train["class"]))
features = list(X_train.columns)

mean_dict = {}
sd_dict = {}
prior_dict = {}

for c in classes:

    X_c = X_train[y_train["class"] == c]

    prior_dict[c] = len(X_c) / len(X_train)

    mean_dict[c] = {}
    sd_dict[c] = {}

    for f in features:
        mean_dict[c][f] = statistics.mean(X_c[f])
        sd_dict[c][f] = statistics.stdev(X_c[f])
```

## Your Code to Create the y_pred

```python
y_pred = []

for i in range(len(X_test)):
    x = X_test.iloc[i]  # one test row (Series)

    best_class = None
    best_score = None

    for c in classes:
        score = np.log(prior_dict[c])

        for f in features:
            mu = mean_dict[c][f]
            sd = sd_dict[c][f]

            p = normal_dist(x[f], mu, sd)
            if p == 0:
                p = 1e-300

            score += np.log(p)

        if (best_score is None) or (score > best_score):
            best_score = score
            best_class = c

    y_pred.append(best_class)
```

## Confusion Matrix

```python
#Confusion Matrix:
cm = confusion_matrix(y_test, y_pred) # <- here, y_pred
disp = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['setosa', 'versicolor', 'virginica'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```
