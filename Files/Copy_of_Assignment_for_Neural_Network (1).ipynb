```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Load dataset
dry_bean = fetch_ucirepo(id=602)
X = dry_bean.data.features
y = dry_bean.data.targets

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=1
)

# Model
mlp = MLPClassifier(
    hidden_layer_sizes=(100,),
    activation='relu',          # changed from logistic
    alpha=0.0001,
    learning_rate_init=0.001,
    max_iter=300,
    random_state=1,
    verbose=True
)

# Training
mlp.fit(X_train, y_train.values.ravel())

# Prediction
y_pred = mlp.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100:.2f}%")
```

Results:
The best accuracy was 94.01 percent. Before scaling, the model accuracy was only around 25 percent, as in the example lab. I applied StandardScaler to normalize feature values since neural networks are sensitive to feature scale. I also switched the activation from logistic to relu which improved training by avoiding vanishing gradient issues. Some moderate alpha values improved the generalization. Furthermore, increasing neurons  improved the accuracy, but deeper networks not so much. Also, having multiple hidden layers and proper scaling also helped lead to the best performance.
